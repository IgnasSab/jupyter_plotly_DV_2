```python
cat_df = df[(df['categories'] == 'Pornstar __Big Tits') | (df['categories'] == 'POV __Amateur') | (df['categories'] == 'MILF __Amateur') | (df['categories'] == 'Big Tits') | (df['categories'] == 'Webcam __Amateur')]
cat_df_gb = cat_df.groupby('categories').sum()
cat_df_gb
```

```python
fig1 = px.bar(data_frame=cat_df_gb, x = cat_df_gb.index, y = 'nb_views')
```

```python

fig2 = px.scatter(data_frame=cat_df[:150], x = 'length', y = 'nb_views', color="categories")
```

```python
df_sort = df[df['nb_views'] < 10000].sort_values(by = 'voting', ascending=True).filter(['nb_views', 'length', 'voting']).reset_index()
df_sort
```

```python
fig3 = px.histogram(data_frame = df_sort[:300], x = 'voting', y = 'nb_views', nbins = 65, range_x = [55, 90])
```

```python
fig1.write_html('first_figure.html', auto_open=True)
```

```python
# Saving to files

fig1.write_html('p-fig-1.html', auto_open = True)
fig2.write_html('p-fig-2.html', auto_open = True)
fig3.write_html('p-fig-3.html', auto_open = True)
```

```python
with open('p_graph.html', 'a') as f:
    f.write(fig1.to_html(full_html=False, include_plotlyjs='cdn'))
    f.write(fig2.to_html(full_html=False, include_plotlyjs='cdn'))
    f.write(fig3.to_html(full_html=False, include_plotlyjs='cdn'))
```
